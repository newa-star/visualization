```python
import altair as alt
import pandas as pd

# Load the  dataset
df = pd.read_csv('tmdb_5000_movies.csv')
```

```python
#See genres
genres = set()
for genre_list in df['genres']:
    genre_dicts = eval(genre_list)
    for genre_dict in genre_dicts:
        genres.add(genre_dict['name'])

genres
```

```python
#see data type
df.dtypes
# The dataset has the following features:-

# budget - The budget in which the movie was made.
# genre - The genre of the movie, Action, Comedy ,Thriller etc.
# homepage - A link to the homepage of the movie.
# id - This is infact the movie_id as in the first dataset.
# keywords - The keywords or tags related to the movie.
# original_language - The language in which the movie was made.
# original_title - The title of the movie before translation or adaptation.
# overview - A brief description of the movie.
# popularity - A numeric quantity specifying the movie popularity.
# production_companies - The production house of the movie.
# production_countries - The country in which it was produced.
# release_date - The date on which it was released.
# revenue - The worldwide revenue generated by the movie.
# runtime - The running time of the movie in minutes.
# status - "Released" or "Rumored".
# tagline - Movie's tagline.
# title - Title of the movie.
# vote_average - average ratings the movie recieved.
# vote_count - the count of votes recieved.
```

```python
import json

# assuming 'df' is the DataFrame containing the 'genres' column
# genre_names = df['genres'].apply(lambda x: [d['name'] for d in json.loads(x)])
# genre_names 
# # options=list(iris['species'].unique()
# list(df['spoken_languages'].unique())

             
# genres_selector = alt.binding_select(options:'genre_names')
# genres_dropdown = alt.selection_single(fields=['genres'], bind=experience_selector, name='GENERS: ') 
df['genres'] = df['genres'].apply(json.loads)

# Extract genre names and create a new column
df['genre_names'] = df['genres'].apply(lambda x: [genre['name'] for genre in x])
df.head()
```

```python
# df['genres'] = df['genres'].apply(json.loads)

# Extract the genre names and create a new column
pd.set_option('display.max_columns', None)  #set to show max rows

df['genre_names'] = df['genres'].apply(lambda x: [genre['name'] for genre in x])

df_exploded = df.explode('genre_names')

# Print the first five rows of the new DataFrame
print(df_exploded.head(100))
df_exploded.shape
```

```python
df['release_date']
```

```python

alt.data_transformers.enable('default', max_rows=None)#change max show restrict(may lead to collapse)
brush = alt.selection_interval(
    encodings=['x'] # limit selection to x-axis (year) values
)

# dynamic query histogram
#This is the first gram to show the year and movie genres on that year
#设计1：考虑到1980前的电影发行量不多，所以时间跨度展示了1980-2016年
#设计2： 图1为柱状图，可以很好地比较出各年份发行的电影的genre的总数量
#设计3：选中图1时，图2的非联动部分会显示出更淡的颜色，这有利于筛选出与图1相关的内容
year_count = alt.Chart(df_exploded).mark_bar(clip=True,color = "#74c476").add_selection(
    brush
).encode(
    alt.X('year(release_date):T', title='movie types by movie release Year',scale=alt.Scale(domain=[1980,2017])),
    alt.Y('count():Q', title=None),

).properties(
    width=650,
    height=50,
)

#popular topic change
#设计4： 采用不同颜色标记电影的genre，让图表更为直观
#设计5： 鼠标放在某个圆圈上时可显示电影的相关信息（主题、预算、收入、流行度、评分），便于查找和信息之间的比对
#设计6： 图二设置了合适的透明度，故可以通过颜色深浅来判断某个genre大致都是什么样的流行度（这句话组织得不好，我不知道怎么表达）
popularity = alt.Chart(df_exploded).mark_circle().encode(
    x='genre_names:N',
    y='popularity:Q',
    size=alt.Size('vote_average:Q', scale=alt.Scale(range=[0,500])),
    color = alt.Color('genre_names:N', scale=alt.Scale(scheme='category20c')),
    tooltip=['original_title:N','budget','revenue','popularity','vote_average'],
    opacity=alt.condition(brush, alt.value(0.4), alt.value(0.05)),
#     shape = alt.Shape('genre_names'),
).properties(
    width=650,
    height=400,
)
#the relationship between budget and revenue
revenue = alt.Chart(df_exploded).mark_circle().encode(
    x='budget:Q',
    y='revenue:Q',
    color = alt.Color('genre_names:N', scale=alt.Scale(scheme='category20c')),
    size=alt.Size( scale=alt.Scale(range=[0,500])),
    tooltip=['original_title:N','genre_names:N','budget','revenue',],
    opacity=alt.condition(brush, alt.value(0.6), alt.value(0.05)),
#     shape = alt.Shape('genre_names'),
).properties(
    width=650,
    height=400,
)
chart = alt.vconcat(year_count, popularity,revenue).properties(
    spacing=10,

           )
chart.save("System A.html")

```
